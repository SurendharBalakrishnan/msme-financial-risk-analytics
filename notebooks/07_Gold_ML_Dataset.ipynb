## Unified ML dataset

```python
ml_dataset = spark.table('msme_risk_analytics.silver_ml_features') \
    .select(
        'loan_amount', 'income', 'Credit_Score', 'LTV', 'dtir1',
        'loan_to_income_ratio', 'risk_score', 'Status'
    ) \
    .dropna()

ml_dataset.write.format('delta').mode('overwrite') \
    .saveAsTable('msme_risk_analytics.gold_ml_training_data')

print(f"✓ Created: gold_ml_training_data ({ml_dataset.count()} rows)")
ml_dataset.describe().show()
```
